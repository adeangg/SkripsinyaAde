```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 12
```

```python
data_skripsi = [90153, 87023, 94190, 97512, 97400, 95103, 96012, 93192, 108479, 118720, 114112, 106970, 110932, 107321, 120994]
timestamp_skripsi = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

hasil_GWOFDDGM =  [
    90153.0000, 87023.0000, 96052.1515, 94046.9658, 95929.1238,
    97889.7732, 99923.9827, 102026.5861, 104194.4672, 106425.9133,
    108720.1163, 111076.8676, 113496.3724, 115979.1334, 118525.8748
]

hasil_GM = [
    90153.0000, 89212.5617, 91363.7199, 93566.7485, 95822.8981,
    98133.4496, 100499.7148, 102923.0371, 105404.7924, 107946.3895,
    110549.2715, 113214.9160, 115944.8365, 118740.5829, 121603.7423
]

hasil_FDGM = [
    90153.0000, 89614.1827, 91490.8606, 93549.9900, 95730.2141,
    98007.3810, 100370.2274, 102813.0635, 105333.0002, 107928.7154,
    110599.8397, 113346.6220, 116169.7365, 119070.1645, 122049.1196
]
```

```python
chart = pd.DataFrame({
    'Tahun': timestamp_skripsi,
    'Data Asli': data_skripsi,
    'GWO-FDDGM': hasil_GWOFDDGM,
    'GM': hasil_GM,
    'GWO-FDGM': hasil_FDGM
})
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def middle(data):
    return (min(data) + max(data)) / 2

plt.figure(figsize=(10, 6))
plt.plot(chart['Tahun'], chart['Data Asli'], marker='o', label='Data Asli')
plt.plot(chart['Tahun'], chart['GWO-FDDGM'], marker='x', label='GWO-FDDGM')
plt.plot(chart['Tahun'], chart['GM'], marker='s', label='GM')
plt.plot(chart['Tahun'], chart['GWO-FDGM'], marker='^', label='GWO-FDGM')

# Add a vertical line to separate the areas
plt.axvline(x=chart['Tahun'][n-1], color='black', linestyle='--', label='Pemisah Simulasi/Prediksi')

# Add text annotations for simulasi and prediksi areas
plt.text(chart['Tahun'][6], max(chart['Data Asli']), 'Simulasi', ha='left', va='center', weight='semibold')
plt.text(chart['Tahun'].iloc[-2], middle(chart['Data Asli']), 'Prediksi', ha='center', va='center', weight='semibold')


plt.xlabel('Tahun')
plt.ylabel('Nilai (satuan ribuan BoE)')
plt.title('Grafik Simulasi dan Prediksi pada Model Grey Berbeda')
plt.legend()
plt.grid(True)
plt.show()
```
